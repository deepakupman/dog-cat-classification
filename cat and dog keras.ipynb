```
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout
```

```
model = Sequential()
# convo layer 1
model.add(Conv2D(16, kernel_size=3, padding='same', activation='relu', input_shape=(64,64,3)))
model.add(MaxPool2D(2))
# convo layer 2
model.add(Conv2D(32, kernel_size=3, padding='same', activation='relu'))
model.add(MaxPool2D(2))
# # convo layer 3
model.add(Conv2D(64, kernel_size=3, padding='same', activation='relu'))
# max pooling layer
model.add(MaxPool2D(2))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```
from keras.preprocessing.image import ImageDataGenerator
```

**Load train and test data set**

```
training_dataset = ImageDataGenerator(rescale=1./255,
                                      shear_range=0.2,
                                      zoom_range=0.2,
                                      horizontal_flip=True)
test_dataset = ImageDataGenerator(rescale=1./255)
```

**Load training set**

```
training_set = training_dataset.flow_from_directory("../input/data-for-dogs-and-cats/data/data/train",
                                                   target_size=(64,64),
                                                   batch_size=32,
                                                   class_mode='binary')
```

**Load test set**

```
test_set = training_dataset.flow_from_directory("../input/data-for-dogs-and-cats/data/data/validation",
                                                   target_size=(64,64),
                                                   batch_size=32,
                                                   class_mode='binary')
```

**Training data**

```
model.fit_generator(training_set,
                    epochs=1,
                    validation_data=test_set,
                    use_multiprocessing=True,
                    steps_per_epoch=20000,
                    validation_steps=5000)
```
